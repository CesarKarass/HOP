# [dbo].[CambiarPassword]

```sql
/****** Object:  StoredProcedure [dbo].[CambiarPassword]    Script Date: 19/05/2026 10:11:29 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[CambiarPassword]
    @UsuarioID INT,
    @PasswordActual NVARCHAR(255),
    @NuevaPassword NVARCHAR(255)
AS
BEGIN
    SET NOCOUNT ON;
    
    DECLARE @PasswordEnBD NVARCHAR(255);
    
    -- Obtener la contraseña actual del usuario
    SELECT @PasswordEnBD = Password FROM Usuarios WHERE Id = @UsuarioID;
    
    -- Verificar si la contraseña actual es correcta
    IF @PasswordEnBD IS NULL OR @PasswordEnBD = 'GOOGLE_AUTH_USER'
    BEGIN
        SELECT 0 AS Exito, 'No puedes cambiar la contraseña porque usas inicio de sesión con Google' AS Error;
        RETURN;
    END
    
    IF @PasswordEnBD != @PasswordActual
    BEGIN
        SELECT 0 AS Exito, 'La contraseña actual es incorrecta' AS Error;
        RETURN;
    END
    
    -- Actualizar la contraseña
    UPDATE Usuarios 
    SET Password = @NuevaPassword
    WHERE Id = @UsuarioID;
    
    SELECT 1 AS Exito, 'Contraseña actualizada correctamente' AS Mensaje;
END
GO
```

# [dbo].[CrearCalificacion]

```sql
/****** Object:  StoredProcedure [dbo].[CrearCalificacion]    Script Date: 19/05/2026 10:11:29 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[CrearCalificacion]
    @ServicioID INT,
    @CalificadorID INT,
    @CalificadoID INT,
    @Puntuacion INT,
    @Comentario NVARCHAR(500) = NULL
AS
BEGIN
    SET NOCOUNT ON;
    
    -- Verificar si ya existe calificación de este usuario para este servicio
    IF EXISTS (SELECT 1 FROM Calificaciones WHERE ServicioID = @ServicioID AND CalificadorID = @CalificadorID)
    BEGIN
        SELECT 'Ya has calificado este servicio' AS Error, 0 AS Success;
        RETURN;
    END
    
    -- Verificar que el usuario no califique su propio servicio
    IF @CalificadorID = @CalificadoID
    BEGIN
        SELECT 'No puedes calificar tu propio servicio' AS Error, 0 AS Success;
        RETURN;
    END
    
    -- Verificar que la puntuación sea válida
    IF @Puntuacion < 1 OR @Puntuacion > 5
    BEGIN
        SELECT 'La puntuación debe ser entre 1 y 5' AS Error, 0 AS Success;
        RETURN;
    END
    
    -- Insertar la calificación
    INSERT INTO Calificaciones (ServicioID, CalificadorID, CalificadoID, Puntuacion, Comentario, Fecha, Leida)
    VALUES (@ServicioID, @CalificadorID, @CalificadoID, @Puntuacion, @Comentario, GETDATE(), 0);
    
    DECLARE @CalificacionId INT = SCOPE_IDENTITY();
    
    -- Crear notificación para el calificado
    INSERT INTO Notificaciones (UsuarioID, Tipo, Contenido, ReferenciaID, FechaCreacion, Leida)
    VALUES (@CalificadoID, 'calificacion', 'Has recibido una nueva calificación', @CalificacionId, GETDATE(), 0);
    
    SELECT 'Calificación creada exitosamente' AS Mensaje, @CalificacionId AS CalificacionId, 1 AS Success;
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[EliminarUsuario]    Script Date: 19/05/2026 10:11:29 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Eliminar usuario (los datos relacionados se eliminarán por ON DELETE CASCADE)
CREATE   PROCEDURE [dbo].[EliminarUsuario]
    @UsuarioId INT
AS
BEGIN
    -- Primero eliminar postulaciones
    DELETE FROM Postulaciones WHERE PrestadorID = @UsuarioId;
    
    -- Eliminar notificaciones
    DELETE FROM Notificaciones WHERE UsuarioID = @UsuarioId;
    
    -- Eliminar mensajes
    DELETE FROM Mensajes WHERE EmisorID = @UsuarioId OR ReceptorID = @UsuarioId;
    
    -- Eliminar conversaciones
    DELETE FROM Conversaciones WHERE Usuario1ID = @UsuarioId OR Usuario2ID = @UsuarioId;
    
    -- Eliminar calificaciones
    DELETE FROM Calificaciones WHERE CalificadorID = @UsuarioId OR CalificadoID = @UsuarioId;
    
    -- Eliminar CVs
    DELETE FROM UsuarioCV WHERE UsuarioId = @UsuarioId;
    
    -- Eliminar servicios
    DELETE FROM Servicios WHERE UsuarioID = @UsuarioId;
    
    -- Finalmente eliminar usuario
    DELETE FROM Usuarios WHERE Id = @UsuarioId
END
GO
```

```sql
/****** Object:  StoredProcedure [dbo].[ObtenerConversacion]    Script Date: 19/05/2026 10:11:29 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Procedimiento para obtener conversación
CREATE   PROCEDURE [dbo].[ObtenerConversacion]
    @User1 INT,
    @User2 INT
AS
BEGIN
    DECLARE @ConvId INT
    
    SELECT @ConvId = Id 
    FROM Conversaciones 
    WHERE (Usuario1ID = @User1 AND Usuario2ID = @User2)
       OR (Usuario1ID = @User2 AND Usuario2ID = @User1)
    
    IF @ConvId IS NULL
    BEGIN
        INSERT INTO Conversaciones (Usuario1ID, Usuario2ID, FechaInicio)
        VALUES (@User1, @User2, GETDATE())
        
        SET @ConvId = SCOPE_IDENTITY()
    END
    
    SELECT @ConvId AS id
END
GO
```

# [dbo].[ObtenerEstadisticasCalificaciones]

```sql
/****** Object:  StoredProcedure [dbo].[ObtenerEstadisticasCalificaciones]    Script Date: 19/05/2026 10:11:29 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE   PROCEDURE [dbo].[ObtenerEstadisticasCalificaciones]
    @UsuarioID INT
AS
BEGIN
    SELECT 
        COUNT(*) AS TotalCalificaciones,
        ISNULL(AVG(CAST(Puntuacion AS FLOAT)), 0) AS Promedio,
        SUM(CASE WHEN Puntuacion = 5 THEN 1 ELSE 0 END) AS Estrellas5,
        SUM(CASE WHEN Puntuacion = 4 THEN 1 ELSE 0 END) AS Estrellas4,
        SUM(CASE WHEN Puntuacion = 3 THEN 1 ELSE 0 END) AS Estrellas3,
        SUM(CASE WHEN Puntuacion = 2 THEN 1 ELSE 0 END) AS Estrellas2,
        SUM(CASE WHEN Puntuacion = 1 THEN 1 ELSE 0 END) AS Estrellas1
    FROM Calificaciones
    WHERE CalificadoID = @UsuarioID
END
GO
```

# [dbo].[ObtenerNotificaciones]

```sql
/****** Object:  StoredProcedure [dbo].[ObtenerNotificaciones]    Script Date: 19/05/2026 10:11:29 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Modificar ObtenerNotificaciones para incluir postulaciones
CREATE   PROCEDURE [dbo].[ObtenerNotificaciones]
    @UsuarioID INT
AS
BEGIN
    -- Notificaciones de mensajes
    SELECT 
        m.Id,
        'Mensaje de ' + u.Nombre AS Contenido,
        'mensaje' AS Tipo,
        ISNULL(m.Leido, 0) AS Leida,
        m.FechaEnvio AS FechaCreacion,
        m.ConversacionID AS ReferenciaId,
        m.EmisorID AS EmisorId,
        u.Nombre AS EmisorNombre,
        m.Mensaje AS MensajeContenido,
        NULL AS Puntuacion,
        NULL AS ComentarioCalificacion,
        NULL AS CalificadorNombre,
        NULL AS ServicioTitulo
    FROM Mensajes m
    INNER JOIN Usuarios u ON m.EmisorID = u.Id
    WHERE m.ReceptorID = @UsuarioID
    
    UNION ALL
    
    -- Notificaciones de calificaciones
    SELECT 
        c.Id,
        'Calificación de ' + u.Nombre AS Contenido,
        'calificacion' AS Tipo,
        ISNULL(c.Leida, 0) AS Leida,
        c.Fecha AS FechaCreacion,
        c.ServicioID AS ReferenciaId,
        c.CalificadorID AS EmisorId,
        u.Nombre AS EmisorNombre,
        NULL AS MensajeContenido,
        c.Puntuacion,
        c.Comentario AS ComentarioCalificacion,
        u.Nombre AS CalificadorNombre,
        s.Titulo AS ServicioTitulo
    FROM Calificaciones c
    INNER JOIN Usuarios u ON c.CalificadorID = u.Id
    INNER JOIN Servicios s ON c.ServicioID = s.Id
    WHERE c.CalificadoID = @UsuarioID
    
    UNION ALL
    
    -- Notificaciones de postulaciones
    SELECT 
        p.Id,
        'Postulación de ' + u.Nombre + ' para tu servicio "' + s.Titulo + '"' AS Contenido,
        'postulacion' AS Tipo,
        0 AS Leida,
        p.FechaPostulacion AS FechaCreacion,
        p.Id AS ReferenciaId,
        p.PrestadorID AS EmisorId,
        u.Nombre AS EmisorNombre,
        p.Mensaje AS MensajeContenido,
        NULL AS Puntuacion,
        NULL AS ComentarioCalificacion,
        NULL AS CalificadorNombre,
        s.Titulo AS ServicioTitulo
    FROM Postulaciones p
    INNER JOIN Usuarios u ON p.PrestadorID = u.Id
    INNER JOIN Servicios s ON p.ServicioID = s.Id
    WHERE s.UsuarioID = @UsuarioID
    
    ORDER BY FechaCreacion DESC
END
GO
```

# [dbo].[UpsertUsuarioGoogle]

```sql
/****** Object:  StoredProcedure [dbo].[UpsertUsuarioGoogle]    Script Date: 19/05/2026 10:11:29 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Crear el procedimiento actualizado
CREATE   PROCEDURE [dbo].[UpsertUsuarioGoogle]
    @Email NVARCHAR(100),
    @Nombre NVARCHAR(100),
    @Telefono NVARCHAR(20) = NULL,
    @Ubicacion NVARCHAR(100) = NULL
AS
BEGIN
    SET NOCOUNT ON;

    DECLARE @UsuarioId INT;
    DECLARE @RolID INT = 2; -- Rol de usuario normal

    -- Verificar si el usuario ya existe por Email
    SELECT @UsuarioId = Id FROM Usuarios WHERE Email = @Email;

    IF @UsuarioId IS NULL
    BEGIN
        -- Insertar nuevo usuario
        INSERT INTO Usuarios (
            Nombre, 
            NombreCompleto, 
            Email, 
            RolID, 
            Activo, 
            FechaRegistro
        )
        VALUES (
            @Nombre,
            @Nombre,
            @Email,
            @RolID,
            1,
            GETDATE()
        );
        
        SET @UsuarioId = SCOPE_IDENTITY();
        
        PRINT 'Usuario insertado: ' + CAST(@UsuarioId AS NVARCHAR);
    END
    ELSE
    BEGIN
        PRINT 'Usuario ya existe: ' + CAST(@UsuarioId AS NVARCHAR);
    END

    -- Retornar la información del usuario
    SELECT 
        u.Id,
        u.Nombre AS [name],
        u.Email,
        r.Nombre AS [Rol],
        ISNULL(u.NombreCompleto, u.Nombre) AS NombreCompleto
    FROM Usuarios u
    INNER JOIN Roles r ON u.RolID = r.Id
    WHERE u.Id = @UsuarioId;
END
GO
```
